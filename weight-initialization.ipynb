```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load a suitable dataset (e.g., MNIST)
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 28*28).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28*28).astype('float32') / 255.0
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# Define a function to create the neural network model with different weight initializations
def create_model(init_type):
    model = keras.Sequential()
    model.add(layers.InputLayer(input_shape=(28*28,)))
    model.add(layers.Dense(256, activation='relu', kernel_initializer=init_type))
    model.add(layers.Dense(128, activation='relu', kernel_initializer=init_type))
    model.add(layers.Dense(10, activation='softmax'))
    return model

# List of weight initialization techniques to compare
weight_initializations = ['zeros', 'random_normal', 'glorot_normal', 'he_normal']

# Train and evaluate models with different weight initializations
results = {}
for init_type in weight_initializations:
    model = create_model(init_type)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test))
    results[init_type] = history

# Compare the performance of initialized models
for init_type, history in results.items():
    print(f"Weight Initialization: {init_type}")
    print("Training Accuracy:", max(history.history['accuracy']))
    print("Validation Accuracy:", max(history.history['val_accuracy']))
    print("-" * 50)
```
